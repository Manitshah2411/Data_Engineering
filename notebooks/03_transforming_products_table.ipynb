```python
import pandas as pd

df_cust = pd.read_csv('../data/raw/csv/customers.csv')
df_pro = pd.read_csv('../data/raw/csv/products.csv')
df_ord = pd.read_csv('../data/raw/csv/orders.csv')
df_ordit = pd.read_csv('../data/raw/csv/order_items.csv')
```

```python
df_pro.info()
```

```python
# if the columns which are supposed to be numeric but are string are needed to be converted into numeric columns

num_cols = ['product_name_lenght',
            'product_description_lenght',
            'product_photos_qty',
            'product_weight_g',
            'product_length_cm',
            'product_height_cm',
            'product_width_cm']

for col in num_cols:
    df_pro[col] = pd.to_numeric(df_pro[col], errors='coerce')
    
print(df_pro[col].dtypes) # as all the columns were already numeric this steps are not needed it is just for information
```

```python
# checking zeros or negatives present in the columns where it shouldn't be and replacing it with NA

for col in num_cols:
    df_pro.loc[df_pro[col] <= 0,col] = pd.NA 
```

```python
# calculating the volumne as per the given columns (l * h * b)

df_pro['product_volume_cm3'] = (
    df_pro['product_length_cm'] *
    df_pro['product_height_cm'] *
    df_pro['product_width_cm']
)
```

```python
df_pro['product_category_name'] = df_pro['product_category_name'].fillna('unknown')

cols = ['product_name_lenght','product_description_lenght','product_photos_qty','product_weight_g','product_length_cm','product_height_cm',
        'product_width_cm','product_volume_cm3']


df_pro[cols] = df_pro[cols].fillna(0)

df_pro[cols].isna().sum()
df_pro.isna().sum()    
```

```python
# asserting some test cases

assert df_pro['product_id'].is_unique, 'duplicate found'
assert (df_pro['product_photos_qty'] >= 0).all()
```
